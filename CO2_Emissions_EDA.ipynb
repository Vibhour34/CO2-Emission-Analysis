# **DATASET :**

## This dataset contains official CO2 emissions data for various car of different features over the period of 2014 to 2020. It has 7385 samples and total 11 features.
## There are few abbreviations that has been used in this data for different features. The abbreviations are listed below :
## **1. Model :**
#### **4WD/4X4** = Four-wheel drive
#### **AWD** = All-wheel drive
#### **FFV** = Flexible-fuel vehicle
#### **SWB** = Short wheelbase
#### **LWB** = Long wheelbase
#### **EWB** = Extended wheelbase
## **2. Transmission :**
#### **A** = automatic
#### **AM** = automated manual
#### **AS** = automatic with select shift
#### **AV** = continuously variable
#### **M** = manual
## **3. Fuel type :**
#### **X** = regular gasoline
#### **Z** = premium gasoline
#### **D** = diesel
#### **E** = ethanol (E85)
#### **N** = natural gas
## **4. Fuel consumption :**
#### City and highway fuel consumption ratings are shown in litres per 100 kilometres (L/100 km) - the combined rating (55% city, 45% hwy) is shown in L/100 km and in miles per imperial gallon (mpg)

## ***Split Transmission column into Transimission and No. of Gears***

```python
df["Gears"] = df["Transmission"].str.extract(r'(\d+)')
```

```python
df["Gears"] = df["Gears"].fillna(0).astype(int)
```

```python
df["Transmission"] = df["Transmission"].str.replace(r'\d+', '', regex=True).str.strip()
```

## ***Descriptive Statistical Analysis***

```python
numeric_df = df.select_dtypes(include=["number"])
stats_ = numeric_df.describe().T.drop(["count"], axis=1)
stats_ = pd.concat([stats_, numeric_df.skew()], axis=1)
stats_.columns = ["mean", "std", "min", "25%", "median", "75%", "max", "skew"]
cols = ["mean", "25%", "median", "75%", "std", "skew", "min", "max"]
stats_ = stats_[cols]
print(tabulate(stats_, headers="keys", floatfmt=".2f"))
```

### Statistical Analysis Summary

### The range of CO₂ emissions lies between 96 g/km and 522 g/km.

### The mean CO₂ emissions is 250.58 g/km, while the median is 246 g/km, indicating a fairly balanced distribution.

### 25% of vehicles have CO₂ emissions below 208 g/km, whereas 75% of vehicles emit below 288 g/km.

### Most features shows positive skewness, suggesting the presence of higher extreme values.

### Cylinders and Fuel Consumption Hwy (L/100 km) show the highest skewness and CO₂ Emissions (g/km) is the least skewed feature among all variables.

## ***CO2 Emission anlaysis against diffrent factors***

### ***Against Categorical Column***

### 1. Brand

```python
brand = df.groupby(["Brand"])["CO2 Emissions(g/km)"].mean().sort_values().reset_index()

# plt.figure(figsize=(20,5))
brand.plot(x="Brand", y="CO2 Emissions(g/km)",kind="bar", figsize=(20,5), legend=False)
plt.title('CO2 Emissions variation with Brand', fontsize=15)
plt.xlabel('Brand', fontsize=12)
plt.xticks(rotation=90, horizontalalignment='center')
plt.ylabel('CO2 Emissions(g/km)', fontsize=12)
plt.show()
```

```python
brand = df.groupby("Brand")["CO2 Emissions(g/km)"].median().sort_values(ascending=True).index

data = [df[df["Brand"] == brand_]["CO2 Emissions(g/km)"] for brand_ in brand]

plt.figure(figsize=(20,7))
plt.boxplot(data, tick_labels=brand, widths=0.5, patch_artist=True)

plt.title("Distribution of CO2 Emissions in relation to Brand", fontsize=15)
plt.xticks(rotation=90, ha='center')
plt.xlabel("Brand", fontsize=12)
plt.ylabel("CO2 Emissions (g/km)", fontsize=12)

plt.axhline(df["CO2 Emissions(g/km)"].median(), color='r', linestyle='dashed', linewidth=2)

plt.show()
```

### Bar plot and Boxplot both show that Smart Brand has the lowest and Bugatti brand has the highest average CO2 emission.

### The red dotted line in boxplot shows the sample median of CO2 emission.

### Among all the brands, 50% of the brands have their median of CO2 emission higher and 50% of the brands have their median of CO2 emission lower from sample median of CO2 emission.

### AUDI, Chrysler and Lexus have their median of CO2 emission almost same from Sample median of CO2 emission.

## --------------------------------------------------------------------------------------------------------------

### 6. Fuel Consumption(City, Hwy, Comb)

```python
plt.figure(figsize=(18,5))

# City
plt.subplot(1,3,1)
plt.scatter(df["Fuel Consumption City (L/100 km)"], df["CO2 Emissions(g/km)"], color="green")
plt.xlabel("Fuel Consumption City (L/100 km)")
plt.ylabel("CO2 Emissions (g/km)")
plt.title("City vs CO2 Emissions")

# Highway
plt.subplot(1,3,2)
plt.scatter(df["Fuel Consumption Hwy (L/100 km)"], df["CO2 Emissions(g/km)"], color="orange")
plt.xlabel("Fuel Consumption Hwy (L/100 km)")
plt.ylabel("CO2 Emissions (g/km)")
plt.title("Highway vs CO2 Emissions")

# Combined
plt.subplot(1,3,3)
plt.scatter(df["Fuel Consumption Comb (L/100 km)"], df["CO2 Emissions(g/km)"], color="blue")
plt.xlabel("Fuel Consumption Comb (L/100 km)")
plt.ylabel("CO2 Emissions (g/km)")
plt.title("Combined vs CO2 Emissions")

plt.tight_layout()
plt.show()
```

### All three scatter plots show a strong positive linear relationship: higher fuel consumption leads to higher CO2 emissions.

### City fuel consumption is the highest, resulting in comparatively higher emissions.

### Highway fuel consumption is lower, leading to relatively reduced emissions.

### Combined consumption lies between city and highway values and provides a balanced summary of emissions.
